```python
import pandas as pd
import folium
```

```python
# criando um df com os dados do gps
dados_gps = pd.read_csv('GPS2LOG5.CSV')
print(dados_gps.shape) # checando a quantidade de elementos do df
dados_gps.head(5) 

```

```python
gps_location = folium.Map(
    zoom_start=30,
    location=[dados_gps["latitude"][2],dados_gps["longitude"][2]] # local que vai aparecer no mapa
)

# criando os pontos
for _, i in dados_gps.iterrows():
    folium.CircleMarker(
        location=[i['latitude'], i['longitude']], radius=0.01
    ).add_to(gps_location)

gps_location
```
